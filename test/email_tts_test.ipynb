```python
# --- Install required packages ---
!pip install gTTS

import os
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from gtts import gTTS
from google.colab import files
from IPython.display import Audio

# --- 1. Upload or Define Transcript ---
print("üìÇ Upload your transcript.txt file or skip to use inline text.")
uploaded = files.upload()

if uploaded:
    transcript_file = list(uploaded.keys())[0]
    with open(transcript_file, 'r', encoding='utf-8') as f:
        transcript_text = f.read()
else:
    # Fallback default text
    transcript_text = "Hello! This is a test transcript for email and TTS."

print("\nTranscript loaded:\n", transcript_text[:200], "...")


# --- 2. Set Email Credentials (Live, in Colab) ---
# ‚ö† Use an App Password for Gmail, not your main password
os.environ['EMAIL_ADDRESS'] = input("Enter your email address: ").strip()
os.environ['EMAIL_PASSWORD'] = input("Enter your email password or App Password: ").strip()

SENDER = os.environ['EMAIL_ADDRESS']
PASSWORD = os.environ['EMAIL_PASSWORD']
RECIPIENT = input("Enter recipient email: ").strip()


# --- 3. Function to Send Email ---
def email_transcript(sender_email, sender_password, recipient_email, subject, transcript_text):
    try:
        msg = MIMEMultipart()
        msg["From"] = sender_email
        msg["To"] = recipient_email
        msg["Subject"] = subject
        msg.attach(MIMEText(transcript_text, "plain"))

        server = smtplib.SMTP("smtp.gmail.com", 587)
        server.starttls()
        server.login(sender_email, sender_password)
        server.send_message(msg)
        server.quit()

        print(f"‚úÖ Transcript emailed to {recipient_email}")
    except Exception as e:
        print(f"‚ùå Failed to send email: {e}")


# --- 4. Send Transcript as Email ---
email_transcript(SENDER, PASSWORD, RECIPIENT, "Colab Transcript Test", transcript_text)


# --- 5. Convert Transcript to TTS ---
tts = gTTS(text=transcript_text, lang='en')
tts_output = "transcript_tts.mp3"
tts.save(tts_output)
print(f"‚úÖ TTS audio saved as {tts_output}")

# --- 6. Play Audio in Colab ---
Audio(tts_output)

# --- 7. Download MP3 ---
files.download(tts_output)
```
