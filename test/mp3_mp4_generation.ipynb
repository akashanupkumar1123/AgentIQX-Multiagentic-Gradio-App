```python
# ===============================
# 1. Install gTTS
# ===============================
!pip install gTTS --quiet

# ===============================
# 2. Import libraries
# ===============================
from gtts import gTTS
import os
from IPython.display import Audio
from google.colab import files

# ===============================
# 3. Prepare the text
# ===============================
speech_text = """
Hello there! This is a sample audio file created for testing purposes.
I am speaking clearly so that transcription models, such as Whisper, can easily understand my words.
This recording will last for at least twenty seconds to provide enough material for testing.
You can use this audio to test how well a model can convert speech into text, regardless of background noise or accent.
The goal is to have a clean, clear voice for accurate transcription in your application.
Thank you for listening, and good luck with your testing process.
"""

# ===============================
# 4. Generate the audio
# ===============================
tts = gTTS(text=speech_text, lang='en', slow=False)

# Save locally in Colab runtime
mp3_filename = "sample_audio.mp3"
tts.save(mp3_filename)

print(f"✅ MP3 generated and saved as: {mp3_filename}")

# ===============================
# 5. Play the audio in Colab
# ===============================
display(Audio(filename=mp3_filename))

# ===============================
# 6. Download the file to your machine
# ===============================
files.download(mp3_filename)
```

```python
# ======================================================
# 1. Install Dependencies
# ======================================================
!pip install gTTS moviepy --quiet

from gtts import gTTS
from moviepy.editor import ColorClip, AudioFileClip
from IPython.display import Audio as IPyAudio
from google.colab import files
import os

# ======================================================
# 2. Prepare Speech Text (~20+ seconds)
# ======================================================
speech_text = """
Hello there! This is a sample audio file created for testing purposes.
I am speaking clearly so that transcription models, such as Whisper, can easily understand my words.
This recording will last for at least twenty seconds to provide enough material for testing.
You can use this audio to test how well a model can convert speech into text, regardless of background noise or accent.
The goal is to have a clean, clear voice for accurate transcription in your application.
Thank you for listening, and good luck with your testing process.
"""

# ======================================================
# 3. Generate the MP3 Audio
# ======================================================
mp3_filename = "sample_audio.mp3"
tts = gTTS(text=speech_text, lang='en', slow=False)
tts.save(mp3_filename)
print(f"✅ MP3 generated: {mp3_filename}")

# Preview in Colab
display(IPyAudio(filename=mp3_filename))

# ======================================================
# 4. Create MP4 Video with Audio
# ======================================================
# A simple color clip (e.g., white background)
audio_clip = AudioFileClip(mp3_filename)
duration = audio_clip.duration  # video duration matches audio

video_clip = ColorClip(size=(1280, 720), color=(255, 255, 255))  # white bg
video_clip = video_clip.set_duration(duration)
video_clip = video_clip.set_audio(audio_clip)

mp4_filename = "sample_video.mp4"
video_clip.write_videofile(mp4_filename, fps=24)

print(f"✅ MP4 generated: {mp4_filename}")

# ======================================================
# 5. Download Files to Local Machine
# ======================================================
print("⬇️ Download MP3")
files.download(mp3_filename)

print("⬇️ Download MP4")
files.download(mp4_filename)
```
